```python
import chess
import random
from collections import Counter


stats = list()
turn_stats = list()

for i in range(100000):
    board = chess.Board()
    turn_counter = 0
    while True:
        turn_counter += 1
        move = random.choice(list(board.legal_moves))
        board.push(move)
        if board.is_game_over():
            break
    stats.append(board.result())
    turn_stats.append(turn_counter)

stats_counter = dict(Counter(stats))
board
```

```python
import numpy as np
import matplotlib.pyplot as plt

turn_stats_2 = [x/2 for x in turn_stats]
counts, bins = np.histogram(turn_stats_2, bins=1000)
plt.hist(bins[:-1], bins, weights=counts)
```
